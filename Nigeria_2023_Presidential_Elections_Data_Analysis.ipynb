```python
politically_active = elections_frame.loc[:, [("State", "State"), ("TurnoutPercentage", "TurnoutPercentage")]]
```

```python
politically_active
```

```python
most_turnout = politically_active.loc[:, ("TurnoutPercentage", "TurnoutPercentage")].max()
```

```python
most_politically_active = politically_active.loc[:, ("TurnoutPercentage", "TurnoutPercentage")] == most_turnout
```

```python
politically_active[most_politically_active][("State", "State")]
```

```python
least_turnout = politically_active.loc[:, ("TurnoutPercentage", "TurnoutPercentage")].min()
```

```python
least_politically_active = politically_active.loc[:, ("TurnoutPercentage", "TurnoutPercentage")] == least_turnout
```

```python
politically_active[least_politically_active][("State", "State")]
```

```python
politically_active.head(10)
```

```python
sort_pol_active = politically_active.sort_values(by=[("TurnoutPercentage", "TurnoutPercentage")])
```

```python
sort_pol_active
```

```python
y = sort_pol_active.loc[:, ("State", "State")]
x = sort_pol_active.loc[:, ("TurnoutPercentage", "TurnoutPercentage")]
```

```python
x
```

```python
len(y)
```

```python
plt.title("Political Awareness and Activity")
plt.xlabel("Turnout Percentage(%)")
plt.ylabel("States")
x_ticks = np.arange(0, 150, 2)
y_ticks = np.arange(0, 150, 5)
plt.yticks(x_ticks)
plt.xticks(y_ticks)
plt.margins(0.02)
#plt.subplots_adjust(bottom=0.15)
plt.grid(color="lightgray", alpha=0.5, zorder=1)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.bar(x, y, color="lightgreen", alpha=0.9, zorder=2, width=0.68)

plt.show()
```
